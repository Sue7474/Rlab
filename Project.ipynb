```python
import pandas as pd
import matplotlib.pyplot as plt 
from tkinter import Tk, Label, Button, OptionMenu, StringVar, Radiobutton, IntVar


file_path = '/Users/sukhalisinghaulakh/Downloads/Breast_Cancerf.csv'  
data = pd.read_csv(file_path)

root = Tk()
root.title("Data Visualization Dashboard")
root.geometry("400x300")

selected_column = StringVar()
plot_type = IntVar()

Label(root, text="Select Column:").pack()
column_dropdown = OptionMenu(root, selected_column, *data.columns)
column_dropdown.pack()

Label(root, text="Select Plot Type:").pack()

plot_options = [
    ("Histogram", 1),
    ("Bar Plot", 2),
    ("Line Plot", 3),
    ("Pie Chart", 4),
    ("Area Plot", 5)
]

for text, value in plot_options:
    Radiobutton(root, text=text, variable=plot_type, value=value).pack(anchor='w')

def generate_plot():
    column = selected_column.get()
    if column == "":
        print("Please select a column.")
        return

    plot_choice = plot_type.get()

    if plot_choice == 1:
        plt.hist(data[column], bins=20, color='red', edgecolor='black')
        plt.title(f'Histogram of {column}')
        plt.xlabel(column)
        plt.ylabel('Frequency')

    elif plot_choice == 2:  
        value_counts = data[column].value_counts()
        plt.bar(value_counts.index, value_counts.values, color='brown')
        plt.title(f'Bar Plot of {column}')
        plt.xlabel(column)
        plt.ylabel('Counts')

    elif plot_choice == 3:
        plt.plot(data[column], color='coral')
        plt.title(f'Line Plot of {column}')
        plt.xlabel('Index')
        plt.ylabel(column)

    elif plot_choice == 4:
        value_counts = data[column].value_counts()
        plt.pie(value_counts, labels=value_counts.index, autopct='%1.1f%%', startangle=140)
        plt.title(f'Pie Chart of {column}')

    elif plot_choice == 5:
        plt.fill_between(range(len(data)), data[column], color="green", alpha=0.5)
        plt.title(f'Area Plot of {column}')
        plt.xlabel('Index')
        plt.ylabel(column)

    else:
        print("Invalid plot selection.")
        return

    plt.show()

Button(root, text="Generate Plot", command=generate_plot).pack(pady=10)

root.mainloop()
```
